```python
parts = mecab.parse(text)
parts
```

```python
nouns = []
for part in parts.split('\n')[:-2]:
  if '名詞' in part.split('\t')[4]:
    nouns.append(part.split('\t')[0])
```

```python
nouns
```

```python
# nounsはリスト型だったので、空白を入れてjoinによってつなげることで文字列型に変更される
words = ' '.join(nouns)
```
