```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
html = urlopen("https://qiita.com/sh1ch/items/b408d524ac3a410f29fc")
data = html.read()
html = data.decode("utf-8")

soup = BeautifulSoup(html, "html.parser")
title = soup.find("title")
body = soup.find("body")
# findAll の場合は複数の要素を返すので、for ループを使用するしかない
a_list = soup.findAll("a", attrs={"class": "style-1ij24kf"})
a_result = []
for a in a_list:
  a_result.append(a.text)

print("title: " + title.text)
print("body: " + body.text)
print(a_result)
```
